```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
def debug_imshow(image):
  fig = plt.figure(figsize=(10, 6), dpi=100)
  plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
  plt.show()

def debug_imshow2(image1, image2):
  fig, axs = plt.subplots(1, 2, figsize=(5, 3))
  axs[0].imshow(image1, cmap='gray')
  axs[0].set_title('_')
  axs[1].imshow(image2, cmap='gray')
  axs[1].set_title('_')
```

```python
image_path = '../../public/static/art-64pixel-2.png'
img = cv.imread(image_path, 0)

prewitt_kernel_x = np.array(
  [[-1, 0, 1],
  [-1, 0, 1],
  [-1, 0, 1]])

edges_x = cv.filter2D(img, -1,prewitt_kernel_x)

debug_imshow2(img, edges_x)
```

```python
image_path = '../../public/static/art-64pixel-2.png'
img = cv.imread(image_path, 0)

prewitt_kernel_y = np.array(
  [[-1, -1, -1],
  [0, 0, 0],
  [1, 1, 1]])

edges_y = cv.filter2D(img, -1, prewitt_kernel_y)

debug_imshow2(img, edges_y)
```

```python
image_path = '../../public/static/art-64pixel-2.png'
img = cv.imread(image_path, 0)

sobel_kernel_x = np.array([
  [-1, -2, 0, 2, 1],
  [-4, -8, 0, 8, 4],
  [-6, -12, 0, 12, 6],
  [-4, -8, 0, 8, 4],
  [-1, -2, 0, 2, 1]
])

sobel_x = cv.filter2D(img, -1, sobel_kernel_x)

debug_imshow2(img, sobel_x)
```
